```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samplerate = 2 ** 10
timestep =  1e-6
tcol = np.linspace(0.0, timestep * samplerate, samplerate, endpoint=False)
```

```python
# Create Random Binary signal

# prbs1: Pseudo Random multilevel Digital signal
numberofpointspersymbol = 2 ** 5
repetitions1 = samplerate / numberofpointspersymbol

rand_int1 = np.random.randint(0, 2, numberofpointspersymbol) # 0 or 1
print(np.size(rand_int1))
print('rand_int1: ', rand_int1)

fig1 = plt.figure(figsize = (8,4), facecolor='lightblue')
ax1_1 = fig1.add_subplot(2, 1, 1)
ax1_1.set_title('Pseudo Random Binary Symbol 1')
ax1_1.stem(rand_int1, ".-", label = 'rand_int1')
ax1_1.legend()

prbs1 = np.repeat(rand_int1, repetitions1)
print(np.size(prbs1))
print('prbs1: ', prbs1)

fig2 = plt.figure(figsize = (8,4), facecolor='lightblue')
ax2_1 = fig2.add_subplot(2, 1, 1)
ax2_1.set_title('Pseudo Random Binary Signal 1')
ax2_1.plot(tcol, prbs1, ".-")
```
