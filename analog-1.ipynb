```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

samplerate =  2 ** 10 # The number of samples in the resampled signal

timestep = 1/ (2 ** 30) 
tcol = np.linspace(0.0, timestep * samplerate, samplerate, endpoint=False)

# Create Random Analog signal
a1_range = [0, 1] #amplitude range
a1 = np.random.rand(samplerate) * (a1_range[1]-a1_range[0]) + a1_range[0] 

b1_range = [48, 128] # frequency range
b1 = np.random.rand(samplerate) *(b1_range[1]-b1_range[0]) + b1_range[0]
b1 = np.round(b1)
b1 = b1.astype(int)

b1[0] = 0
for i in range(1,np.size(b1)):
    b1[i] = b1[i-1]+b1[i]

i=0

random_signal = np.zeros(samplerate, dtype=float)
while b1[i]<np.size(random_signal):
    k = b1[i]
    random_signal[k:] = a1[i]
    i=i+1

#---
fig1 = plt.figure(figsize = (7,3), facecolor='white')
ax1_1 = fig1.add_subplot(1, 1, 1)

ax1_1.set_title("Pseudo Random Analog Signal")
ax1_1.plot(tcol,random_signal, ".-")
ax1_1.set_ylim([-0.1, 1.1])
```
