```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

samplerate =  2 ** 10 # The number of samples in the resampled signal

timestep = 1/ (2 ** 30) 
tcol = np.linspace(0.0, timestep * samplerate, samplerate, endpoint=False)

# Create Random Analog signal
a1_range = [0, 1] #amplitude range
a1 = np.round((np.random.rand(samplerate) * (a1_range[1]-a1_range[0]) + a1_range[0])*255) 
print(np.size(a1))

#---
fig1 = plt.figure(figsize = (7,3), facecolor='white')
ax1_1 = fig1.add_subplot(1, 1, 1)

ax1_1.set_title("Pseudo Random 8-bit Signal")
ax1_1.plot(a1, ".-")




b1_range = [2**5, 128] # frequency range
b1 = np.random.rand(samplerate) *(b1_range[1]-b1_range[0]) + b1_range[0]
b1 = np.round(b1)
b1 = b1.astype(int)
print(np.size(b1))

#---
fig2 = plt.figure(figsize = (7,3), facecolor='white')
ax2_1 = fig2.add_subplot(1, 1, 1)

ax2_1.set_title("Pseudo Random 8-bit Signal")
ax2_1.plot(b1, ".-")



c1_range = np.zeros(int(samplerate / b1_range[0]))
print(np.size(c1_range))


b1[0] = 0
for i in range(1,np.size(b1)):
    b1[i] = b1[i-1]+b1[i]

i=0

random_signal = np.zeros(samplerate, dtype=float)
while b1[i]<np.size(random_signal):
    k = b1[i]
    random_signal[k:] = a1[i]
    i=i+1




fig3 = plt.figure(figsize = (7,3), facecolor='white')
ax3_1 = fig3.add_subplot(1, 1, 1)

ax3_1.set_title("Pseudo Random 8-bit Signal")
ax3_1.plot(tcol,random_signal, ".-")
#ax1_1.set_ylim([-0.1, 1.1])
```

```python
binary_representation = bin(random_signal)

binary_digits = binary_representation[2:]

print(f"The binary digits of {random_signal} are {binary_digits}")
```
